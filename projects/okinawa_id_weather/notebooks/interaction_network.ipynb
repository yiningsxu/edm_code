```python
%matplotlib inline
import networkx as nx  # NetworkXをインポート
import matplotlib.pyplot as plt
import csv
```

```python
G = nx.DiGraph()
t    ["flu", "rs", "pcf", "gas", "aig", "cpox", "hfmd", "eryt", "es", "pert", "herp", "ahc", "ekc", "bm", "am", "mp",
     "cp",
     "aig_rota"
     ])

def csv_to_list(csv_file_path):
    result = []
    # 读取文件时，要注意文件的编码格式与实际情况是否一致（如 UTF-8 等）
    with open(csv_file_path, 'r', encoding='utf-8') as f:
        # 使用 DictReader 可以通过表头来读取指定的字段
        reader = csv.DictReader(f)
        for row in reader:
            cause = row["cause"].strip('"')
            effected = row["effected"].strip('"')
            color = row["color"].strip('"')
            weight = int(row["weight"])
            tp_abs_str = f"[{str(abs(int(row['tp'])))} w]"
            result.append( (cause, effected, color, weight, tp_abs_str) )

    return result

if __name__ == "__main__":
    csv_file = "/Users/ayo/Downloads/significant_res_summary_final_Yaeyama.csv"
    edges = csv_to_list(csv_file)
    print(edges)

for u, v, color, weight, label in edges:
    G.add_edge(u, v, color=color, weight=weight, label=label)

# 提取边属性
colors = [data["color"] for _, _, data in G.edges(data=True)]
weights = [data["weight"] for _, _, data in G.edges(data=True)]
labels = nx.get_edge_attributes(G, "label")

# 可视化设置
pos = nx.circular_layout(G)
plt.figure(figsize=(20, 20), dpi=300)

# 绘制节点和边
nx.draw_networkx(
    G, pos,
    edge_color=colors,
    width=weights,
    with_labels=True,
    node_size=800,
    node_color='lightgreen'
)

# 绘制边标签
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels=labels,
    font_color='red',
    font_size=8
)

plt.axis('off')
plt.savefig("/Users/yining/Desktop/_GSAIS_/mzmtlab/microbiome dynamics/result/oknw_ID/2025-01-17/okinawaID_network_240117.png",
            format="png", dpi=300, bbox_inches="tight")
plt.show()
```
